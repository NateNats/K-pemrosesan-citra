```python
from PIL import Image
import matplotlib.pyplot as plt
import cv2
import numpy as np

img = cv2.imread('daun.jpeg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
```

```python
adaptive_thresh = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)

plt.imshow(adaptive_thresh, cmap='gray')
plt.axis('off')
```

```python
def iterative_threshold(matrix):
    threshold = np.mean(matrix)
    newThreshold = 0
    while abs(threshold - newThreshold)> 1e-5:
        lower = np.mean(matrix[matrix < threshold])
        upper = np.mean(matrix[matrix >= threshold])
        newThreshold = (lower + upper) / 2
        threshold = newThreshold
    return threshold
```

```python
matrix = np.array([
    [165, 250, 55, 145, 145,],
    [178, 68, 60, 25, 67],
    [60, 80, 62, 180, 186],
    [55, 145, 60, 66, 69],
    [68, 60, 55, 50, 145]
])

print(iterative_threshold(matrix))
```
